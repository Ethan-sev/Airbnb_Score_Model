```python
# Generate our categorical variable lists
categorical_features = Airbnb_df.dtypes[Airbnb_df.dtypes == "object"].index.tolist()
```

```python
# Check the number of unique values in each column
Airbnb_df[categorical_features].nunique()
```

```python
numeric_features = X.select_dtypes(include='number')
```

```python
numeric_features.nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)
```

```python
# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(Airbnb_df[categorical_features]))
```

```python
# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names_out(categorical_features)
encode_df.head()
```

```python
# Merge one-hot encoded features and drop the originals
Airbnb_df = Airbnb_df.merge(encode_df,left_index=True, right_index=True)
Airbnb_df = Airbnb_df.drop(columns=categorical_features)
Airbnb_df.head()
```
